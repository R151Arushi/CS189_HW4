```python
import numpy as np
import matplotlib.pyplot as plt

def lp_norm(X, Y, p):
    return (np.abs(X)**p + np.abs(Y)**p)**(1/p)

x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)


X, Y = np.meshgrid(x, y)

#(a)
Diagram = lp_norm(X, Y, 0.5)
plt.contour(X, Y, Diagram)
plt.show()

# (b)
Diagram= lp_norm(X, Y, 1)
plt.contour(X, Y, Diagram)
plt.show()

#(c)
Diagram = lp_norm(X, Y, 2)
plt.contour(X, Y, Diagram)
plt.show()
```
